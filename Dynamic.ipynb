```python
import pandas as pd
```

```python
cflist = pd.read_csv('List.csv')
netf = pd.read_csv('netf.csv')
```

```python
cflist
```

```python
#### customer 12345 has two emails: bobmar@mail.com and bobby@ymail.mail used on two brands

cflist.loc[(cflist['ID'] == 12345)]
```

```python
netf
```

```python
ABC_R = netf.loc[netf['brand']=='ABC', 'returns'].sum()

### sums the returns column for each value which corresponds to ABC in row
```

```python
ABC_S = netf.loc[netf['brand']=='ABC', 'settled'].sum()

### sums the settled column for each value which corresponds to ABC in row
```

```python

### conditional statement only if returns - settlements == 0
if ABC_R - ABC_S == 0:
    
### drops from cflist only for both ID == netf ID and if brand == ABC and reason == returns

    cflist.drop(cflist.loc[(cflist['ID'] == netf['ID'].iloc[0])
                           & (cflist['brand'] == 'ABC') 
                           & (cflist['reason'] == 'returns')].index, inplace= True)
    
print(cflist)
```

```python
### same method used for brand XYZ - can be done per brand 

XYZ_R = netf.loc[netf['brand']=='XYZ', 'returns'].sum()

XYZ_S = netf.loc[netf['brand']=='XYZ', 'settled'].sum()

print(XYZ_R)
print(XYZ_S)
```

```python
if XYZ_R - XYZ_S == 0:
    cflist.drop(cflist.loc[(cflist['ID'] == netf['ID'].iloc[0]) 
                           
                           & (cflist['brand'] == 'XYZ') 
                           
                           & (cflist['reason'] == 'returns')].index, inplace= True)
    
print(cflist)
```
