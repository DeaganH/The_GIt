```python
import IPython.display as display
from PIL import Image 
import cv2
import matplotlib.pyplot as plt
import os
```

```python
cat_1 = "Abyssinian_86.jpg"
cat_2 = "Abyssinian_86_2.jpg"

image_1 = cv2.imread(cat_1)
image_2 = cv2.imread(cat_2)
```

```python
image_1 ### cv2.imread converts the image into a 3D array/tensor
```

```python
### simple comparison of two images
if image_1.shape == image_2.shape:
    print('The images have same size and channels')
    
###subtract the pixel values in two images and merge them
difference = cv2.subtract(image_1, image_2)
### split() is used to split an image into three different intensity arrays for each color
b, g, r = cv2.split(difference)
    
###countNonZero counts the number of non zero pixels 
###the logic is, if they are identical, there should be no non zero pixels remaining
if cv2.countNonZero(b) == 0 and cv2.countNonZero(g) == 0 and cv2.countNonZero(r) == 0:
    print('The images are completely equal')
    
    os.rename(cat_2, 'duplicate image.jpg')
```
